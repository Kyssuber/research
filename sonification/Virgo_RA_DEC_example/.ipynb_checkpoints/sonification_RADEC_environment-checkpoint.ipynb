### preparing data to sonify

```python
homedir=os.getenv("HOME")

#need vf_main (v2) and env (v2) catalogs
vf_cat = Table.read(homedir+'/Desktop/v2-20220820/vf_v2_main.fits')
env_cat = Table.read(homedir+'/Desktop/v2-20220820/vf_v2_environment.fits')
```

```python
ra = vf_cat['RA']
dec = vf_cat['DEC']
radius = vf_cat['radius'] #arcsec
```

```python
pgflag = env_cat['poor_group_memb']
rgflag = env_cat['rich_group_memb']
fieldflag = env_cat['pure_field']
filflag = env_cat['filament_member']
clusflag = env_cat['cluster_member']

filandrg = (filflag) & (rgflag)
filandpg = (filflag) & (pgflag)
filandclus = (filflag) & (clusflag)
```

```python
#my 'y-data' will be environment flags, so convert flags to unique integers
base_list = np.zeros(len(env_cat))
flag_list = [fieldflag,pgflag,filflag,rgflag,clusflag]
int_list = [1,3,5,10,20]     
#I choose wonky numbers to prevent flag ambiguity (e.g., what if 5 is both rg and pg+fil?)
flag_int_list = []

for n in range(len(int_list)):
    flag = flag_list[n]
    integer = int_list[n]
    int_flag = np.zeros(len(env_cat))
    for i in range(len(env_cat)):
        if flag[i] == True:
            int_flag[i]+=integer
    flag_int_list.append(int_flag)
```

```python
env_list = flag_int_list[0]+flag_int_list[1]+flag_int_list[2]+flag_int_list[3]+flag_int_list[4]
```

```python
#I am not too keen on the fil+cluster flag, so I'll set those values to 20 (cluster)
for i in range(len(env_list)):
    if int(env_list[i]) == 25:
        env_list[i] = 20
```

###### KEY:

### introducing the sonification component of this exercise
##### https://medium.com/@astromattrusso/sonification-101-how-to-convert-data-into-music-with-python-71a6dd67751c

```python
#load data as a pandas dataframe
filename='ra_dec_env.csv'
df = pd.read_csv(homedir+'/sonification/Virgo_RA_DEC/'+filename)
```

```python
#a typical sonification mapping function
#mapping value(s) from one range to another range
#returns float(s)

def map_value(value, min_value, max_value, min_result, max_result):
    result = min_result + (value - min_value)/(max_value - min_value)*(max_result - min_result)
    return result
```

```python
#now. compress time. how many galaxies per beat of music?

'''FIRST OPTION: SET CONVERSION FACTOR TO SUPPRESS "TIME"
(we have # galaxies per beat, but what will be the beat duration?)'''

gals_per_beat = 10
#rescale strip number to beats
t_data = df['RA'] / gals_per_beat
t_data = t_data - np.min(t_data)
duration_beats=np.max(t_data)
print('Duration:',duration_beats, 'beats')
#one beat = one quarter note
```

```python
#normalize env_flag data
y_data = df['env_flag']
y_data = map_value(y_data,min(y_data),max(y_data),0.1,1)
```

```python
#visualizing data...
plt.figure(figsize=(6,4))
plt.scatter(t_data,y_data,s=y_data*100,alpha=0.5,edgecolors='black')
plt.xlabel('scaled RA (deg)',fontsize=12)
plt.ylabel('normalized environmental flag',fontsize=12)
plt.show()
```

```python
note_names = ['C2','F2','C3','F3','C4','F4','C5','F5','C6','F6']

note_midis = [str2midi(n) for n in note_names] #make a list of midi note numbers

n_notes = len(note_midis)
print('Resolution:',n_notes,'notes')
```

```python
#MAPPING DATA TO MIDIS!

midi_data = []
#for every data point (which I represent with len(t_data))
for i in range(len(t_data)):
    #map y_data values such that smallest flag is lowest note, largest flag is highest
    #assigns midi note number to whichever y_data[i], once mapped, is nearest to
    note_index = round(map_value(y_data[i],0.1,1,0,len(note_midis)-1))
    midi_data.append(note_midis[note_index])
```

```python
plt.scatter(t_data,midi_data,s=y_data*100,alpha=0.5,edgecolors='black')
plt.xlabel('RA scaled (deg)',fontsize=12)
plt.ylabel('Midi note numbers',fontsize=12)
plt.show()
```

```python
#map data to note velocities
#like midi numbers, ranges from 0 to 127
#don't want vel_min=0, as this value will generate a *very* faint sound
vel_min,vel_max = 15,127
note_velocity = round(map_value(df['radius'],np.min(df['radius']),np.max(df['radius']),vel_min,vel_max))
```

```python
bpm = 30

#create midi file object, add tempo
my_midi_file = MIDIFile(1) #one track
#time = np.min(t_data)     #eliminates time delay...sort of.
time=0
my_midi_file.addTempo(track=0,time=time,tempo=bpm)

#add midi notes to file
for i in range(len(t_data)):
    my_midi_file.addNote(track=0,channel=0,pitch=midi_data[i],time=t_data[i],duration=1,volume=int(note_velocity[i]))
    
#write file
with open('virgo_env1.mid',"wb") as f:
    my_midi_file.writeFile(f)
```
