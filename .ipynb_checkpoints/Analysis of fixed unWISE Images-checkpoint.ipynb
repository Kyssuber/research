## Generating histogram distributions of galfit output parameters to find differences between galaxy fits images corrected and uncorrected for oversubtraction haloes. There appeared to be a few mismatches between corresponding parameters (potentially due to the variable background present in the corrected images), so this visualization will allow a proper glimpse at how significant are the mismatches on larger scale.

```python
uncorrected_59 = ascii.read(homedir+'/github/WISE_test/testest2.txt')
corrected_59 = ascii.read(homedir+'/github/unwise_59_fixed/test_no_psf_cornerplots.txt')
```

```python
flag= np.where(uncorrected_59['mag'] != -999)
prob_flag = [24,30]
print(corrected_59[24]['prefix'],corrected_59[30]['prefix'])
uncorrected_59 = uncorrected_59[flag[0]]
corrected_59 = corrected_59[flag[0]]
#indices with problematic galaxies: 24,30
```

------------


## Effective Radius ($R_e$) Histogram

```python
re_ratio = (corrected_59['re']-uncorrected_59['re'])/uncorrected_59['re']
plt.figure(figsize=(10,8))
plt.hist(re_ratio,label='57/59 Galaxy Subsample')
plt.xlabel(r'$\Delta R_e /R_e$',fontsize=20)
plt.ylabel('# Galaxies',fontsize=20)

plt.legend(loc='upper right',fontsize=20)
plt.title('Distribution of R_e',fontsize=25)
plt.text(4,38,'Mean: ' + str('%0.2f')%np.mean(re_ratio),fontsize=20)
plt.text(4,35,'Median:' +str('%0.2f')%np.median(re_ratio),fontsize=20)
```

```python
# "outlier" galaxy indices:
```

```python
np.where(re_ratio>3)
```

------------

## Magnitude (mag) Histogram

```python
deltamag = corrected_59['mag']-uncorrected_59['mag']
plt.figure(figsize=(10,8))
plt.hist(deltamag,label='59 Galaxy Subsample')
plt.xlabel(r'$\Delta mag$'+' (corr - uncorr)',fontsize=20)
plt.ylabel('# Galaxies',fontsize=20)

plt.legend(loc='upper left',fontsize=20)
plt.title('Distribution of Magnitude',fontsize=25)
plt.text(-2.5,33,'Mean: ' + str('%0.2f')%np.mean(deltamag),fontsize=20)
plt.text(-2.5,31,'Median:' +str('%0.2f')%np.median(deltamag),fontsize=20)
```

```python
# "outlier" galaxy indices:
```

```python
np.where(deltamag<-1.0)
```

---------

## Sersic Number (nsersic) Histogram

```python
nser_ratio = (corrected_59['nsersic']-uncorrected_59['nsersic'])/uncorrected_59['nsersic']
plt.figure(figsize=(10,8))
plt.hist(nser_ratio,label='57/59 Galaxy Subsample')
plt.xlabel(r'$\Delta$'+'nser/nser',fontsize=20)
plt.ylabel('# Galaxies',fontsize=20)

plt.legend(loc='upper right',fontsize=20)
plt.title('Distribution of Sersic Number',fontsize=25)
plt.text(0.75,20,'Mean: ' + str('%0.2f')%np.mean(nser_ratio),fontsize=20)
plt.text(0.75,18,'Median:' +str('%0.2f')%np.median(nser_ratio),fontsize=20)
```

```python
# "outlier" galaxy indices:
```

```python
np.where(nser_ratio < 0.00)
```

```python
np.where(nser_ratio > 0.75)
```
