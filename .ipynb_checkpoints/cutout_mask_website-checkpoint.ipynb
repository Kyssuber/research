```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from astropy.io import ascii
from astropy.io import fits
from astropy.table import Table
import os
from astropy.wcs import WCS


import wget
import matplotlib.image as mpimg

from scipy.stats import scoreatpercentile
from astropy.visualization import simple_norm
```

```python
homedir = os.getenv("HOME")
vf = Table.read(homedir+'/github/vfcut.fits')
```

```python
cutoutdir = homedir+'/github/cutouts/'
maskdir = homedir+'/github/masks/'
```

# Option one: my inefficient approach (pixscale ratios)

```python
#first, an example case...VFID0001
#WISE IMAGE/MASK

image_wise, h = fits.getdata(cutoutdir+'unwise-VFID0001-w3-img-m-trim.fits',header=True)
image_mask, h_mask = fits.getdata(maskdir+'unwise-VFID0001-w3-img-m-mask-trim.fits',header=True)

#convert header information to wcs coordinates
image_wcs = WCS(h)
```

```python
#LS IMAGE

ra = vf[0]['RA']
dec = vf[0]['DEC']
galname = vf[0]['DEC']
#selected initial pixscale is arbitrary
pixscale=.3
image_url ='https://www.legacysurvey.org/viewer/cutout.fits?ra='+str(ra)+'&dec='+str(dec)+'&layer=ls-dr9&pixscale='+str(pixscale)

#LS fits cutout image will be saved here
#create LS_cutouts directory first
filename = homedir+'/github/LS_cutouts/VFID0001'

#if exists, remove and replace
if os.path.exists(filename):
    os.remove(filename)
    image = wget.download(image_url,out=filename)
else:
    image = wget.download(image_url,out=filename)
    
#define header information
image_LS, h_LS = fits.getdata(filename,header=True)
```

```python
#the following loop will identify the pixscale that aligns the images

os.chdir(homedir+'/github/LS_cutouts')
for scale in np.arange(0.05,10,0.02):
    ratio = (h['NAXIS1']*h['CD1_1'])/(h_LS['NAXIS1']*h_LS['CD1_1'])
    if (ratio > 0.97) & (ratio < 1.03):
        pixscale_list.append(scale)
        print(scale)
        break
    else:
        image_url ='https://www.legacysurvey.org/viewer/cutout.fits?ra='+str(ra)+'&dec='+str(dec)+'&layer=ls-dr9&pixscale='+str(scale)
        if os.path.exists(filename):
            os.remove(filename)
            image = wget.download(image_url,out=filename)
        image_LS, h_LS = fits.getdata(filename,header=True)
        #print('ratio',ratio)
        #print('scale',scale)

wcs_LS = WCS(h_LS)
```

```python
#once the above is set, pull the jpg image
#scale=.55
image_url = 'https://www.legacysurvey.org/viewer/cutout.jpg?ra='+str(ra)+'&dec='+str(dec)+'&layer=ls-dr9&pixscale='+str(scale)

ls_jpg = wget.download(image_url)

#change filename to a more comprehensible version
image_desired = 'VFID0001.jpg'
os.rename(ls_jpg,image_desired)
ls_jpg = homedir+'/github/LS_cutouts/'+image_desired
```

```python
#annnd, plotting.
os.chdir(homedir+'/github/LS_cutouts')
images = [image_wise,image_LS,ls_jpg]
titles = ['WISE Image','LS fits','LS Image']

plt.figure(figsize=(14,6))
plt.subplots_adjust(wspace=.0)
for i,im in enumerate(images):
    plt.xlabel('RA')
    if i == 0:
        plt.subplot(1,3,i+1,projection=image_wcs)
        norm = simple_norm(im,stretch='asinh',max_percent=99.5)
        plt.imshow(im,origin='lower',cmap='viridis',norm=norm)
        plt.ylabel('DEC')
        ax = plt.gca()
        ax.set_yticks([])
    if i == 1:
        plt.subplot(1,3,i+1,projection=wcs_LS[1])
        os.chdir(homedir+'/github')
        plt.imshow(mpimg.imread(ls_jpg),origin='lower')
        plt.gca().invert_yaxis()
        plt.axis("off")
        
    plt.title(titles[i],fontsize=16)    
    
#lastly, save image
#!mkdir LS_mosaics
os.chdir('LS_mosaics')
filename = homedir+'/github/LS_mosaics/VFID0001-mosaic.png'
if os.path.exists(filename):
    os.remove(filename)
    plt.savefig('VFID0001-mosaic.png')
else:
    plt.savefig('VFID0001-mosaic.png')
plt.close
```
